```python
import sys
import os
relative_path = 'my_eyes.py'

absolute_path = os.path.abspath(relative_path)
print("Absolute Path:", absolute_path)
sys.path.append(absolute_path)

from my_eyes import CLIENT_ID, CLIENT_SECRET
import pandas as pd
import numpy as np
import spotipy 
from spotipy.oauth2 import SpotifyClientCredentials, SpotifyOAuth
```

```python
relative_path = 'songs.csv'
absolute_path = os.path.abspath(relative_path)
songs = pd.read_csv(absolute_path)
songs.head()
```

```python
client_id = CLIENT_ID
client_secret = CLIENT_SECRET

# Authenticate with Spotify
client_credentials_manager = SpotifyClientCredentials(client_id=client_id, client_secret=client_secret)
sp = spotipy.Spotify(client_credentials_manager=client_credentials_manager)
```

```python
redirect_uri =  "http://localhost"  # Set this in your Spotify Developer Dashboard

# Scope for reading user's playlists
scope = 'playlist-modify-public playlist-modify-private'

# Create SpotifyOAuth instance
sp_oauth = SpotifyOAuth(client_id=client_id, client_secret=client_secret, redirect_uri=redirect_uri, scope=scope)

# Get authorization URL
auth_url = sp_oauth.get_authorize_url()

# Print authorization URL and visit it in your browser
print(f"Please visit this URL to authorize: {auth_url}")

# Get authorization token
code = input("Enter the code from the redirect URI: ")
token_info = sp_oauth.get_access_token(code)

# Create Spotify instance with authorized token
sp = spotipy.Spotify(auth=token_info['access_token'])
```

```python
playlists = sp.current_user_playlists()

# Display playlists
print("Your playlists:")
for playlist in playlists['items']:
    print(f"{playlist['name']} - {playlist['id']}")
```

```python
playlist_id = '7qG1JCsRsQt6Qswi8A4yja'  # Replace with your playlist ID

try:
    playlist = sp.playlist(playlist_id)
    owner_id = playlist['owner']['id']
    print(f"Playlist '{playlist['name']}' is owned by user with ID: {owner_id}")
except Exception as e:
    print(f"Error fetching playlist information: {e}")
```

```python
song_name = songs['Song'].iloc[0]
artist_name = songs['Artist'].iloc[0]
print(song_name)
print(artist_name)

results = sp.search(q=f'track:{song_name} artist:{artist_name}', type='track', limit=1)

if results['tracks']['items']:
    track_uri = results['tracks']['items'][0]['uri']

playlist_id = '7qG1JCsRsQt6Qswi8A4yja'
sp.playlist_add_items(playlist_id, [track_uri])
print(f"Added '{song_name}' by '{artist_name}' to playlist.")
```

```python
for index, row in songs.iterrows():
    song_name = row['Song']
    artist_name = row['Artist']

    # Search for the track on Spotify
    results = sp.search(q=f'track:{song_name} artist:{artist_name}', type='track', limit=1)

    # Check if a track was found
    if results['tracks']['items']:
        track_uri = results['tracks']['items'][0]['uri']

        # Add track to a playlist (replace playlist_id with your playlist ID)
        playlist_id = '7qG1JCsRsQt6Qswi8A4yja'
        sp.playlist_add_items(playlist_id, [track_uri])
        print(f"Added '{song_name}' by '{artist_name}' to playlist.")
    else:
        print(f"Track '{song_name}' by '{artist_name}' not found on Spotify.")
```
